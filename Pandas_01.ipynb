```python
cast = pd.read_csv('cast.csv')
cast.sample(20)
```

Количество ролей в фильме `Inception` с неизвестным значением столбца `n`:

```python
cast[cast['title']=='Inception']['n'].isnull().sum()
```

Количество ролей в фильме `Inception` с известным значением столбца `n`:

```python
cast[cast['title']=='Inception']['n'].notnull().sum()
```

Актеры фильма `North by Northwest` по их "главности", сортируем по значению столбца `n`. Неизвестные значения игнорируются.

```python
cast.sort_values(by=['n'],ascending=False, inplace=True)
cast.dropna(subset=['n'], inplace=True)
cast[cast['title']=='North by Northwest']
```

Количество ролей в немой версии Гамлета 1921 года:

```python
cast[(cast['year']==1921)&(cast['title']=='Hamlet')]['character'].count() #9 ролей
```

Вторые роли Брэда Питта по годам:


```python
cast[(cast['name']=='Brad Pitt')&(cast['n']==2)].filter(items=['year', 'character'])
```
