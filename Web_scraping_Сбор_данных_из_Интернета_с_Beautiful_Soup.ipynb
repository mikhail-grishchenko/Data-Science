# Задание 1

Обратиться к сайту "https://ru.wikipedia.org/wiki/Гарри_Поттер" и посчитать сколько раз встречаются имена каждого из главных героев произведения: Гарри, Гермионы, Рона.

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = "https://ru.wikipedia.org/wiki/%D0%93%D0%B0%D1%80%D1%80%D0%B8_%D0%9F%D0%BE%D1%82%D1%82%D0%B5%D1%80"
r = requests.get(url)  
soup = BeautifulSoup(r.text, "lxml")
```

```python
list_one = re.split("=|>|<|;|,|\n|!", str(soup))   
```

```python
count_harry = 0 
count_hermione = 0 
count_ron = 0 
for element in list_one:
  if "Гарри" in element or "Harry" in element:
    count_harry += 1
  if "Гермиона" in element or "Hermione" in element:
    count_hermione += 1
  if "Рон" in element or "Ron" in element:
    count_ron += 1
print("Счетчик имени 'Гарри':", count_harry)
print("Счетчик имени 'Гермиона':", count_hermione)
print("Счетчик имени 'Рон':", count_ron)
```

# Задание 4

Творческое задание: найти среднюю стоимость детских ботинок на сайте ebay (возьмите первые 10 страниц, чтобы не захламлять трафик)

```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
from time import sleep
import random
```

```python
mylist=[]
mylist_ship=[]
for p in range(1,11):
  url = f"https://ru.ebay.com/b/Boys-Shoes/57929/bn_661841?rt=nc&_pgn={p}"
  r = requests.get(url)  
  soup = BeautifulSoup(r.text, "lxml")

  items = soup.findAll('li', class_='s-item s-item--large s-item--bgcolored')
  for i in items:
    elem = i.find('span', class_='s-item__price').text
    elem = elem.replace(' руб.', '').replace('\xa0', '').replace(' до ', ' ').replace(',', '.').split() 
    for num in elem:
        mylist.append(num)
    try:
      elem_ship = i.find('span', class_='s-item__shipping s-item__logisticsCost').text
      elem_ship = elem_ship.replace('\xa0', '').replace(' руб. за доставку', '').replace('Бесплатная доставка', '0').replace(',', '.')
      mylist_ship.append(elem_ship)
    except:
      pass

  print(f"Обработана страница №{p}")
  sleep(random.randint(5,15))
```

```python
my_df = pd.DataFrame(mylist, columns={'price_shoes'})
```

```python
my_df_ship = pd.DataFrame(mylist_ship, columns={'price_ship'})
```

```python
my_df['price_shoes'] = pd.to_numeric(my_df['price_shoes'])
```

```python
my_df_ship['price_ship'] = pd.to_numeric(my_df_ship['price_ship'])
```

```python
my_df['price_shoes'].median() #средняя стоимость детских ботинок, с 10-ти первых запрошенных страниц, руб
```

```python
my_df_ship['price_ship'].median() #средняя стоимость доставки детских ботинок, с 10-ти первых запрошенных страниц, руб
```
