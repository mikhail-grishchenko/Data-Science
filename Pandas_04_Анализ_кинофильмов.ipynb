№4. Какие русские фильмы оказались в рейтинге IMDB? Какой жанр встречается чаще всего? 


```python
df7 = df1[df1['country'] == 'Russia'][['movie_title', 'genres', 'color', 'title_year', 'director_name', 'movie_imdb_link']]
```

```python
df7 #русские фильмы в рейтинге IMDB
```

```python
s = df7['genres'].str.split('|').apply(Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = 'genres'
del df7['genres']
df8 = df7.join(s)
```

```python
df9 = pd.pivot_table(df8, index = ["genres"], values = ["movie_title"], aggfunc='count')
```

```python
df10 = df9.sort_values(by='movie_title', ascending=False)
df10 #Adventure жанр встречается чаще всего
```
