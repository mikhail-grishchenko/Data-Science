Изобразим график функции по заданным точкам x и y цветом 'magenta'.

Изобразим сетку синим цветом точечной линией толщиной 0.3.

Напишем "легенду" графика, подпишем оси, дадим название графику. 

Изобразим график в диапазоне x от -10 до 10  и y от -10 до 10, 
определим шаг шкалы для каждого второго целого значения (-10, -8, -6 ..).

```python
t = np.linspace(0, 60, 10000)

x=4.4*(np.cos(t)+np.cos(1.1*t)/1.1)
y=4.4*(np.sin(t)-np.sin(1.1*t)/1.1)


plt.figure(figsize=(10, 10))
plt.plot(x,y, color='magenta', label='Hypocycloid')
plt.grid(color='blue', ls='dotted', linewidth = 0.3)
plt.legend(fontsize=18, shadow=True, loc='upper right')
plt.xlabel('Ось x', fontsize=20)
plt.ylabel('Ось y', fontsize=20)
plt.title('График параметрической функции', fontsize=25)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xticks(np.arange(-10, 12, 2), fontsize=12)
plt.yticks(np.arange(-10, 12, 2),fontsize=12)
plt.show()
```

Ееще один интересный график:

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='параметрическая кривая 1', marker='o')
x = r * np.sin(theta + 1)
y = r * np.cos(theta + 1)
ax.plot(x, y, z, label='параметрическая кривая 2', marker='o')
ax.legend(loc='best')
ax.set_xlabel('Оcь X', fontsize = 14)
ax.set_ylabel('Оcь Y', fontsize = 14)
ax.set_zlabel('Оcь Z', fontsize = 14)

plt.show()
```

```python
from numpy.lib.function_base import meshgrid

fig=plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')

x_data = np.arange(-5, 5, 0.01)
y_data = np.arange(-5, 5, 0.01)
x,y = meshgrid(x_data, y_data)
z = np.sin(x) * np.cos(y)

ax.plot_surface(x, y, z, cmap = 'inferno', label='sin(x)')
ax.set_xlabel('Оcь X', fontsize = 14)
ax.set_ylabel('Оcь Y', fontsize = 14)
ax.set_zlabel('Оcь Z', fontsize = 14)

plt.show()
```

Изобразим 4 графика (2 х 2) для функций sin, cos, tan, cot с помощью subplot. (В некоторых точках функции не существуют).

Оформим графики, сделаем подписи, подберем размеры так, чтобы надписи разных графиков не перекрывались.

```python
x = np.linspace(-10, 10, 1000)
x_tan_cot = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
```

```python
fig, ax = plt.subplots(4, figsize=(10,10))

plt.subplots_adjust(wspace=1.5, hspace=1.5)
fig.set_figwidth(15)
fig.set_figheight(15)

ax[0].plot(x, np.sin(x), label='sin(x)')
ax[0].set_xlabel('Оcь X', fontsize = 14)
ax[0].set_ylabel('Оcь Y', fontsize = 14)
ax[0].set_title('График sin(x)', fontsize=25)
ax[0].grid(color='grey', linestyle='--', linewidth=0.5)
ax[0].legend(shadow=True, loc='best')
ax[0].title.set_position([0.5, 1.2])
ax[0].legend(fontsize = 10,
          facecolor = 'oldlace',    
          edgecolor = 'gray',    
          title_fontsize = '10',   
          shadow=True,
          loc='best'
         )

ax[1].plot(x, np.cos(x), label='cos(x)')
ax[1].set_xlabel('Оcь X', fontsize = 14)
ax[1].set_ylabel('Оcь Y', fontsize = 14)
ax[1].set_title('График cos(x)', fontsize=25)
ax[1].grid(color='grey', linestyle='--', linewidth=0.5)
ax[1].legend(shadow=True, loc='best')
ax[1].title.set_position([0.5, 1.2])
ax[1].legend(fontsize = 10,
          facecolor = 'oldlace',    
          edgecolor = 'gray',    
          title_fontsize = '10',    
          shadow=True,
          loc='best'
         )

ax[2].plot(x_tan_cot, np.tan(x_tan_cot), label='tan(x)')
ax[2].set_ylim(-10, 10)
ax[2].set_xlabel('Оcь X', fontsize = 14)
ax[2].set_ylabel('Оcь Y', fontsize = 14)
ax[2].set_title('График tan(x)', fontsize=25)
ax[2].grid(color='grey', linestyle='--', linewidth=0.5)
ax[2].legend(shadow=True, loc='best')
ax[2].title.set_position([0.5, 1.2])
ax[2].legend(fontsize = 10,
          facecolor = 'oldlace',   
          edgecolor = 'gray',    
          title_fontsize = '10',    
          shadow=True,
          loc='best'
         )

ax[3].plot(x_tan_cot, np.cos(x_tan_cot)/np.sin(x_tan_cot), label='cot(x)')
ax[3].set_ylim(-10, 10)
ax[3].set_xlabel('Оcь X', fontsize = 14)
ax[3].set_ylabel('Оcь Y', fontsize = 14)
ax[3].set_title('График cot(x)', fontsize=25)
ax[3].grid(color='grey', linestyle='--', linewidth=0.5)
ax[3].legend(shadow=True, loc='best')
ax[3].title.set_position([0.5, 1.2])
ax[3].legend(fontsize = 10,
          facecolor = 'oldlace',   
          edgecolor = 'gray',    
          title_fontsize = '10',   
          shadow=True,
          loc='best'
         )

plt.show()
```
